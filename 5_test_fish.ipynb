```python
import numpy as np
import pandas as pd
```

```python
fish = pd.read_csv('https://raw.githubusercontent.com/wizard113/ML-Basic/main/fish_data.csv')
```

```python
fish.head()
```

```python
fish.describe()
```

```python
## Data, Target л¶„лҰ¬
```

```python
np.unique(fish['Species'], return_counts=True)
```

```python
fish.columns
```

```python
input= fish[['Weight', 'Length', 'Diagonal', 'Height', 'Width']].to_numpy()
target = fish['Species'].to_numpy()
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
# н‘ңмӨҖнҷ”
ss = StandardScaler()
ss.fit(input)
input_scaled = ss.transform(input)
print(input_scaled[0])
```

```python
## Train, Test л¶„лҰ¬
```

```python
from sklearn.model_selection import train_test_split
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
x_train, x_test, y_train, y_test = train_test_split(input_scaled, target, test_size=0.2, random_state=23)
```

```python
print(x_train.shape, x_test.shape)
print(y_train.shape, y_test.shape)
```

```python
## Model мғқм„ұ
```

```python
from tensorflow.keras import models, layers, utils
```

```python
model = models.Sequential()
model = models.Sequential([
    layers.Dense(16, activation='relu', input_shape=(5,)),
    layers.Dense(8, activation='relu'),
    layers.Dense(7, activation='softmax')  # л¬јкі кё° мў…лҘҳ 7к°ң
])
model.summary()
```

```python
print(y_train.shape, y_test.shape)
```

```python
print(y_train)
```

```python
# л¬ёмһҗм—ҙ вҶ’ мҲ«мһҗ : one-hot encodingмқ„ мң„н•Ё(м—¬кё°м„ңлҠ” мӮ¬мҡ©н•ҳм§Җ м•Ҡм•„лҸ„ лҗЁ)
le = LabelEncoder()
y_train = le.fit_transform(y_train)
y_test = le.transform(y_test)
```

```python
# one-hot encoding : л¶„лҘҳмқҙлҜҖлЎң мӮ¬мҡ©н•ҳм§Җ м•Ҡм•„лҸ„ лҗЁ
y_train_oh = utils.to_categorical(y_train)
y_test_oh = utils.to_categorical(y_test)
print(y_train_oh.shape, y_test_oh.shape)
```

```python
print(y_train_oh[:10])
```

```python
# л¶„лҘҳмқҙлҜҖлЎң sparse_categorical_crossentropy
model.compile(loss='sparse_categorical_crossentropy', optimizer='adam', metrics=['accuracy'])
```

```python
model.fit(x_train, y_train, epochs=100)
```

```python
# лӘЁлҚё нҸүк°Җ
model.evaluate(x_test, y_test)
```

```python
y_test[:10]
```

```python
# мҳҲмёЎ
pred = model.predict(x_test[:10])
print(pred)
```

```python
pred_index = np.argmax(pred, axis=1)
print(pred_index)
```
